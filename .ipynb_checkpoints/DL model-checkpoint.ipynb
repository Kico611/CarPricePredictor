```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler , OrdinalEncoder, OneHotEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input ,  LeakyReLU
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from tensorflow.keras.losses import Huber
from tensorflow.keras.callbacks import ReduceLROnPlateau
```

```python
data = pd.read_csv("data_preprocessed.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data = data.drop(columns=['Popularity'])
```

```python
data.head()
```

```python
categorical_cols = ['Make', 'Engine Fuel Type', 'Transmission Type', 'Driven_Wheels', 'Vehicle Size', 'Vehicle Style']

for col in categorical_cols:
    unique_vals = data[col].nunique()
    print(f"Stupac '{col}' ima {unique_vals} jedinstvenih vrijednosti.")
```

```python
# Prikaz svih unikatnih stilova vozila s brojem pojavljivanja
vehicle_style_counts = data['Vehicle Style'].value_counts()
print(vehicle_style_counts)
```

```python
def simplify_vehicle_style(style):
    if style in ['4dr SUV', '2dr SUV', 'Convertible SUV']:
        return 'SUV'
    elif style in ['2dr Hatchback', '4dr Hatchback']:
        return 'Hatchback'
    elif style in ['Crew Cab Pickup', 'Extended Cab Pickup', 'Regular Cab Pickup']:
        return 'Pickup'
    elif style in ['Passenger Minivan', 'Passenger Van', 'Cargo Van', 'Cargo Minivan']:
        return 'Van'
    elif style in ['Sedan']:
        return 'Sedan'
    elif style in ['Coupe']:
        return 'Coupe'
    elif style in ['Convertible']:
        return 'Convertible'
    elif style in ['Wagon']:
        return 'Wagon'
    else:
        return 'Other'

# Primjena funkcije
data['Vehicle Style'] = data['Vehicle Style'].apply(simplify_vehicle_style)

# Provjera nove raspodjele
print(data['Vehicle Style'].value_counts())
```

```python
# 1. Kopija podataka
df_encoded = data.copy()

# 2. Ordinalne kategorije
ordinal_cat_cols = ['Make', 'Engine Fuel Type', 'Transmission Type', 'Driven_Wheels']
ordinal_enc = OrdinalEncoder()
df_encoded[ordinal_cat_cols] = ordinal_enc.fit_transform(df_encoded[ordinal_cat_cols])

# 3. One-hot enkodiranje za diskretne numeričke (tretirane kao kategorije)
one_hot_cat_cols = ['Engine Cylinders', 'Number of Doors']
df_encoded[one_hot_cat_cols] = df_encoded[one_hot_cat_cols].astype(str)  # Pretvori u string prije one-hot
df_encoded = pd.get_dummies(df_encoded, columns=one_hot_cat_cols)

# 4. One-hot za postojeće kategorijske varijable
df_encoded = pd.get_dummies(df_encoded, columns=['Vehicle Size', 'Vehicle Style'])

# 5. Pretvori bool -> int (nije nužno, ali čini podatke urednijima)
one_hot_cols = [col for col in df_encoded.columns if df_encoded[col].dtype == 'bool']
df_encoded[one_hot_cols] = df_encoded[one_hot_cols].astype(int)

# 6. Skaliraj samo kontinuirane numeričke varijable
numeric_cols = ['Engine HP', 'Vehicle Age', 'average_mpg']
scaler_X = StandardScaler()
df_encoded[numeric_cols] = scaler_X.fit_transform(df_encoded[numeric_cols])

# 7. Ciljna varijabla bez skaliranja (kasnije ćeš možda raditi log-transformaciju)
y = df_encoded['MSRP'].values

# 8. Ulazne varijable
X = df_encoded.drop(columns=['MSRP'])
```

```python
X.head()
```

```python
X.describe()
```

```python
# --- 1. Log-transformacija ciljne varijable ---
y_log = np.log1p(y)  # log(1+y), tako da nema problema s y=0 ili vrlo malim vrijednostima

# --- 2. Podjela podataka na train/test, ali sada s log-transformiranim y ---
X_train, X_test, y_train_log, y_test_log = train_test_split(X, y_log, test_size=0.2, random_state=42)

# --- 3. Definicija modela (kao i prije) ---
model = Sequential([
    Input(shape=(X_train.shape[1],)),

    Dense(64, activation='relu'),
    BatchNormalization(),

    Dense(128, activation='relu'),
    BatchNormalization(),

    Dense(64, activation='relu'),
    BatchNormalization(),

    Dense(32, activation='relu'),
    BatchNormalization(),

    Dense(1)
])

model.compile(optimizer=Adam(learning_rate=0.002),
              loss=Huber(),
              metrics=['mean_absolute_error'])

# --- 4. Treniranje na log(y) ---
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=20,
    verbose=1,
    min_lr=1e-6
)

history = model.fit(
    X_train, y_train_log,
    validation_split=0.2,
    epochs=300,
    batch_size=64,
    verbose=2,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True),
        reduce_lr
    ]
)
# --- 5. Evaluacija na test skupu (još uvijek log-skalirano y) ---
test_loss, test_mae = model.evaluate(X_test, y_test_log, verbose=2)
print(f"Test MAE (log-skala): {test_mae:.4f}")

# --- 6. Predikcija i vraćanje u originalnu skalu ---
y_pred_log = model.predict(X_test).flatten()
y_pred = np.expm1(y_pred_log)        # inverzna transformacija log(1+y) => y
y_true = np.expm1(y_test_log)        # inverzna transformacija stvarnih vrijednosti

# --- 7. Izračun MAE u originalnim jedinicama ---
mae_original = mean_absolute_error(y_true, y_pred)
print(f"MAE (originalne jedinice): {mae_original:.2f}")

# --- 8. Vizualizacija predikcija vs stvarnih vrijednosti ---
plt.figure(figsize=(10,6))
plt.hist(y_true, bins=50, alpha=0.5, label='Stvarne vrijednosti')
plt.hist(y_pred, bins=50, alpha=0.5, label='Predikcije')
plt.legend()
plt.title('Distribucija stvarnih vrijednosti i predikcija (MSRP)')
plt.xlabel('MSRP')
plt.ylabel('Broj uzoraka')
plt.grid(True)
plt.tight_layout()
plt.show()
```
